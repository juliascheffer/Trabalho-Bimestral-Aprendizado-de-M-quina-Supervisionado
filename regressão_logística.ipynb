```python
rep_df = df.replace(["associate's degree", "bachelor's degree", "master's degree"], ["associates degree", "bachelors degree", "masters degree"])
rep_df.head(7)
```

```python
rep_df.shape
```

```python
# Transformando as variáveis categóricas nominais em variáveis categóricas ordinais
df2 = pd.DataFrame.copy(rep_df)
```

```python
df2['gender'].replace({'male':0, 'female': 1}, inplace=True)
df2['race/ethnicity'].replace({'group A':0, 'group B': 1, 'group C': 2, 'group D': 3, 'group E': 4}, inplace=True)
df2['lunch'].replace({'standard':0, 'free/reduced': 1}, inplace=True)
df2['test preparation course'].replace({'completed':0, 'none': 1}, inplace=True)
df2['parental level of education'].replace({'some high school':0, 'high school': 1, 'some college': 2, 
                                           'bachelors degree': 3, 'associates degree': 4, 'masters degree': 5}, inplace=True)
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
# OneHotEncoder: Criação de variáveis Dummy (fictícia)
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
encoder = OneHotEncoder()
oneHotEncoder = encoder.fit_transform(df[['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']])
```

```python
oneHotEncoder.todense()
```

```python
oneHotEncoder.shape
```

```python
nome_colunas = encoder.get_feature_names(['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course'])
df3_encoder = pd.DataFrame(oneHotEncoder.todense(), columns= nome_colunas)
df3_encoder
```

```python
df3 = pd.concat([df, df3_encoder], axis=1).drop(['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course'], axis=1)
df3.info()
```

Legenda:

Gender: Gênero

Race/Ethnicity: Raça ou etnia

Parental Level of Education: Nível de escolaridade dos pais

Luch: Almoço (se era padrão ou de graça/reduzido)

Test Preparation Course: Curso preparatório para testes - sem é completo ou se não tem

Math Score: Pontuação em matemática

Reading Score: Pontuação em leitura

Writing Score: Pontuação em escrita

```python
from sklearn.preprocessing import StandardScaler
```

```python
previsores_esc = StandardScaler().fit_transform(previsores)
```

```python
# BASE DE TREINO E TESTE
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(previsores_esc, alvo, test_size = 0.3, random_state = 0)
```

```python
x_treino.shape
```

```python
y_treino.shape
```

```python
x_teste.shape
```

```python
y_teste.shape
```

Regressão logística

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistica = LogisticRegression(random_state=1, max_iter=500, penalty="l2",
                               tol=0.0001, C=2,solver="lbfgs")
logistica.fit(x_treino, y_treino)
```

```python
logistica.intercept_
```

```python
logistica.coef_
```

```python
previsoes_logistica = logistica.predict(x_teste)
previsoes_logistica
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_logistica) * 100.0))
```

```python
confusion_matrix(y_teste, previsoes_logistica)
```

```python
print(classification_report(y_teste, previsoes_logistica))
```

```python
# Análise dados de treino
previsoes_treino = logistica.predict(x_treino)
previsoes_treino
```

```python
accuracy_score(y_treino, previsoes_treino)
```

```python
confusion_matrix(y_treino, previsoes_treino)
```
